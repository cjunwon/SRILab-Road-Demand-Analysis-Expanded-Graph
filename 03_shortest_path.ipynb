```python
origin_nodes_list = [t[0] for t in Node_to_Node_pairs]
origin_nodes_list = list(dict.fromkeys(origin_nodes_list))
```

```python
random.seed(123)
origin_nodes_list_subset = random.sample(origin_nodes_list, 5)
```

```python
# make Node_to_Node_pairs_subset, which is a list of tuples where t[0] is in origin_nodes_list_subset

Node_to_Node_pairs_subset = [t for t in Node_to_Node_pairs if t[0] in origin_nodes_list_subset]
```

```python
len(Node_to_Node_pairs_subset)
```

```python
# subset Node_to_Node_pairs_dict to include only the keys in origin_nodes_list_subset

Node_to_Node_pairs_dict_subset = {k: v for k, v in Node_to_Node_pairs_dict.items() if k in origin_nodes_list_subset}

# calculate the total number of pairings in Node_to_Node_pairs_dict_subset

total_pairings = 0

for key, value in Node_to_Node_pairs_dict_subset.items():
    total_pairings += len(value)

total_pairings
```

```python
def compute_shortest_path(origin_node, destination_nodes_list, G):
    shortest_paths = nx.single_source_dijkstra_path(G, origin_node, weight='weight')
    shortest_paths_dict = {k: v for k, v in shortest_paths.items() if k in destination_nodes_list}
    return origin_node, shortest_paths_dict
```

```python
Node_to_Node_shortest_paths_dict = {}

for origin_node in origin_nodes_list_subset:
    destination_nodes_list = Node_to_Node_pairs_dict_subset[origin_node]
    origin_node, shortest_paths_dict = compute_shortest_path(origin_node, destination_nodes_list, G)
    Node_to_Node_shortest_paths_dict
```

```python
# Single Source Method

All_shortest_paths_dict = {}
Node_to_Node_shortest_paths_dict = {}

time_start = time.time()

for pairing in Node_to_Node_pairs_subset:
    node_1 = pairing[0]
    node_2 = pairing[1]

    if node_2 in All_shortest_paths_dict and node_1 in All_shortest_paths_dict[node_2]:
        shortest_path = All_shortest_paths_dict[node_2].get(node_1, None)
        if shortest_path is not None:
            shortest_path = shortest_path[::-1] # reverse the list
        Node_to_Node_shortest_paths_dict[pairing] = shortest_path
    elif node_1 in All_shortest_paths_dict:
        shortest_path = All_shortest_paths_dict[node_1].get(node_2, None)
        Node_to_Node_shortest_paths_dict[pairing] = shortest_path
    else:
        All_shortest_paths_dict[node_1] = nx.single_source_dijkstra_path(G, node_1, weight='weight')
        shortest_path = All_shortest_paths_dict[node_1].get(node_2, None)
        Node_to_Node_shortest_paths_dict[pairing] = shortest_path


time_end = time.time()

print("Time taken for Single Source Method: ", (time_end - time_start)/60, "minutes")
```
